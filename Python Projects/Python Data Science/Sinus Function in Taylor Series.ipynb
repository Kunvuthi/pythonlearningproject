```python
import math
import numpy as np
from matplotlib import pyplot as plt

def sin(x,n):
    value = 0
    for i in range(n):
        sign = (-1)**i
        above = x**((2*i)+1)
        bottom = math.factorial((2*i)+1) 
        value += ((sign*above)/bottom)
    return value

ax = plt.subplot()
x_values = np.arange(0,4*math.pi+1,math.pi/16)
y_values = [sin(x1,25) for x1 in x_values]
y_values_actual = [math.sin(x2) for x2 in x_values]

plt.plot(x_values,y_values_actual, label='Actual Sin(x)')
plt.plot(x_values,y_values, label='Taylor Series of Sin(X)')
ax.set_xticks([0, math.pi/2, math.pi, (3*math.pi)/2, 2*math.pi, (5*math.pi)/2, 3*math.pi, (7*math.pi)/2, 4*math.pi])
ax.set_xticklabels(['0', 'π/2','π','3π/2','2π','5π/2','3π','7π/2','4π'])
plt.ylim(-1,1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Graph of Taylor Series for sin(x)')
plt.grid()
plt.savefig('taylor_sinx.png')
plt.show()


    
```
