```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("csv_files/uk_mean_temperature.csv")
print(df.head())
```

```python
sns.kdeplot(data=df, x="temperature_degree", hue="location", shade=True)
```

```python
sns.kdeplot(df["hurn_mean_temp"], shade=True)
sns.kdeplot(df["leuchars_mean_temp"], shade=True)
```

```python
sns.set_style("whitegrid")
sns.violinplot(data=df, x="location", y="temperature_degree")
plt.title("Distribution of UK Temperature 1987")

plt.savefig("vplot_distribution_uk_temp_1987.png")
plt.show()
```
