```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
Displacement = [0, 0.082, 0.240, 0.538, 0.903, 0.961, 1.000]
Time = [i*0.1 for i in range(7)]
velocity = np.gradient(Displacement, Time)

tuple1 = list(zip(Time, Displacement))
tuple2 = list(zip(Time, velocity))

df1 = pd.DataFrame(tuple1, columns = ['Time', 'Displacement']) 
df2 = pd.DataFrame(tuple2, columns = ['Time', 'velocity']) 

```

# Plotting The Graphs 






```python
f = plt.figure(figsize=(12, 8))
gs = f.add_gridspec(1, 2)

with sns.axes_style("darkgrid"):
    sns.set_palette("bright")
    ax = f.add_subplot(gs[0, 0])
    sns.lineplot(x="Time", y="Displacement", data=df1, marker="o")
    ax.set_title("Displacement-Time Graph")

with sns.axes_style("darkgrid"):
    sns.set_palette("bright")
    ax = f.add_subplot(gs[0, 1])
    sns.lineplot(x="Time", y="velocity", data=df2, marker="o")
    ax.set_title("Velocity-Time Graph")
    
plt.savefig("Physics Graph.png")
plt.show()
```
